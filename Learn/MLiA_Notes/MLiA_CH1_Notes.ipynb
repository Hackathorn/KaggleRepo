# Tasks of Machine Learning

### Supervised Learning -> Telling the algorithm what to predict
- <b>Classification</b> -> Predict what class an instance will fall into
- <b>Regression</b> -> Prediction of a numeric value

### Unsupervised Learning -> No label or target value given for data
- <b>Clustering</b> -> Grouping similar items
- <b>Density Estimation</b> -> Find statistical valuse that describe the data

### So with all of the different methods, how do I choose which one to use?

- If you're trying to predict or forecast a target value, then you need to look into supervised learning. If not, unsupervised.
    - If yes, is it a discrete value like A/B/C or Red/Yellow? If so, then classification, else regression
    - If unsupervised, are you trying to fit your data into some discrete groups? If so, and that is all you need then you should look into clustering. Do you need to have some numerical estimate about how well fits in group? Density Estimation Algorithm.
    
### Things to know about your data

- Are the features nominal or continuous? Are there outliers? Missing Values? Are you looking for needle in haystack?
